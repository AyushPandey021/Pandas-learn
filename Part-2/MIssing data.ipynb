```python
import numpy as np 
import pandas as pd

data={
    'A':[1,2,3,np.nan,5],
    'B':[1,2,3,np.nan,5],
    'C':[np.nan,2,np.nan,np.nan,5],
    'D':[1,2,3,np.nan,5],
    'E':[1,2,3,np.nan,np.nan]
}
df = pd.DataFrame(data)
```
