```python
# use for any kind or data in coloum , title and any spefic word and data use in this loop to find it.⭐⭐
def extract_episodes(txt):
  check = False
  data = ""
  for i in txt:
    if i == ')':
        check = False
        return data
   
 
    if check == True: 
      data = data + i
    if i == '(':
# for eps no.
       check = True

# this work to if you are only get episod no so this loop work to fectch only eps no 
# this loop start in ( and close in )



```

```python
df["Episodes"]= df["Title"].apply(extract_episodes)
#  add new coloum and extract all eps in coloum 
```

```python
df["Episodes"]= df["Episodes"].str.replace(" eps","")
# remove eps for  episodes coloum and print only number
```

```python
def extraction_time(txt):
    data = ""
    for i in range(len(txt)):
        if txt[i] == ")":
            # Loop from next character to next 18 characters, or end of string
            for j in range(i+1, min(i+1 + 19, len(txt))):
                data += txt[j]
            return data
    return None  # in case ')' not found
```

```python
df['Time_Stamp'] = df['Title'].apply(extraction_time)
df
```

```python

from dateutil import parser

def get_total_months(time_str):
  try:
    start_str, end_str = time_str.split(" - ")
    start_date = parser.parse(start_str.strip())
    end_date = parser.parse(end_str.strip())
    return (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)
  except Exception:
    return None  # In case of unexpected format

df['Total_Months'] = df['Time_Stamp'].apply(get_total_months)
```
